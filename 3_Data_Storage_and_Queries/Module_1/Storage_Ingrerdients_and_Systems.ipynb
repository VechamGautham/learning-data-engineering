# Storage Hierarchy

Storage in data engineering can be thought of as a **hierarchy with three layers**:

---

## 1. Raw Ingredients  
The foundation of all storage systems.  
- **Physical Components**:  
  - Magnetic disks (HDD)  
  - Solid-state drives (SSD)  
  - Memory (RAM)  

- **Processes**:  
  - Networking  
  - Serialization  
  - Compression  
  - CPU utilization  
  - Caching  

---

## 2. Storage Systems  
Built from raw ingredients. Provide management and interaction with data.  
- Databases (OLTP & OLAP)  
- Distributed storage systems  
- Specialized systems:  
  - Graph databases  
  - Vector databases  
- Caching systems  

---

## 3. Storage Abstractions  
High-level assemblies of storage systems.  
- Data warehouses  
- Data lakes  
- Data lakehouses  

---

---

![Improving Performance](./images/performace_2.png)

---

## üîã Volatile Memory

- **CPU Cache** ‚Üí Ultra-fast memory right on the processor, keeps frequently used data close.

# Networking, Serialization & Compression

When working with storage systems, data doesn‚Äôt stay in one place ‚Äî it moves between **memory, disk, and networks**. To make this efficient, several processes come into play.  

---

## üåê Networking & CPU
Networking connects multiple servers in distributed systems.  
It helps with:  
- Faster reads/writes  
- Data durability  
- High availability  

![Networking](./images/networking.png)

---

## üì¶ Serialization
To store data from **memory to disk** or send it across a network, it must be transformed.  
- **Serialization** ‚Üí Convert in-memory structures into disk-friendly format  
- **Deserialization** ‚Üí Rebuild original memory format  

![Serialization](./images/serialize.png)

---

## üìä Serialization Approaches
- **Row-based** ‚Üí Store data record by record (good for transactions).  
- **Column-based** ‚Üí Store data column by column (good for analytics).  

![Serialization Example](./images/serialization.png)

---

## üìù Serialization Formats
You can choose different file formats for serialized data:  

- **Human-readable**:  
  - CSV ‚Üí Simple, row-based, but error-prone (coz no schema)  
  - XML ‚Üí Legacy, slow to process  
  - JSON ‚Üí Widely used for APIs  

- **Binary**:  
  - Parquet ‚Üí Column-based, efficient for big data  
  - Avro ‚Üí Row-based, schema-driven, supports evolution  

![Serialization Formats](./images/serialization_formats.png)

---

## üóú Compression
Once serialized, data can be **compressed** before saving to disk.  
- Reduces disk space  
- Speeds up queries  
- Cuts down I/O time  

![Compression Flow](./images/compressin_1.png)

Compression works by removing **redundancy and repetition** in data.  
For example, frequently used characters get shorter encodings.  

![Compression Example](./images/compression.png)

---

## üîë Flow Summary
1. **Networking/CPU** ‚Üí Move and coordinate data across servers.  
2. **Serialization** ‚Üí Convert memory data into storable/transmittable format.  
3. **Serialization Formats** ‚Üí Choose (CSV/JSON/XML for readability, Parquet/Avro for efficiency).  
4. **Compression** ‚Üí Save space and improve performance before writing to disk.  

# Cloud Storage Options

As a data engineer, you have three main cloud storage options: **File Storage**, **Block Storage**, and **Object Storage**.  
Each has different strengths and trade-offs depending on your workload.

---

## üìÇ File Storage
Organizes files in a **hierarchical directory tree** (like folders on your laptop).  
Each directory stores metadata such as name, owner, permissions, and location pointers.

- ‚úÖ Easy to use and manage  
- ‚úÖ Great for sharing files across users and machines  
- ‚ùå Lower read/write performance (must track hierarchy)  
- ‚ùå Limited scalability compared to other options  

![File Storage](./images/file_storage.png)  
![File Storage Use Case](./images/file_storage_usecase.png)

---

## üì¶ Block Storage
Divides data into **small fixed-size blocks**, each with a unique ID.  
You can directly update or retrieve specific blocks without rewriting the entire file.

- ‚úÖ High performance & low latency ‚Üí ideal for **transactional workloads (OLTP)**  
- ‚úÖ Supports frequent small reads/writes  
- ‚úÖ Used in **databases, VMs (e.g., Amazon EBS)**  
- ‚ùå Limited scalability (usually a few TBs)  
- ‚ùå Tied closely to compute instances (not independent like object storage)  

![Block Storage](./images/block_storage.png)  
![Block Storage Lookup](./images/block_storage_2.png)  
![Block Storage Use Case](./images/block_storage_usecase.png)

---

## üóÉÔ∏è Object Storage
Stores data as **immutable objects** in a flat structure with unique keys.  
Objects are replicated across nodes and scale almost infinitely.

- ‚úÖ Extremely scalable (petabytes and beyond)  
- ‚úÖ Great for **analytics (OLAP), data lakes, ML pipelines**  
- ‚úÖ High durability via replication  
- ‚ùå Not good for transactional workloads (must rewrite entire object to update)  
- ‚ùå Higher latency for frequent small updates  

![Object Storage](./images/object_storage.png)  
![Object Storage Use Cases](./images/object_storage_use_Cases.png)

---

## ‚öñÔ∏è Trade-offs at a Glance

![Storage Options](./images/storage_option.png)

- **File Storage** ‚Üí simple sharing & management, but not for performance-heavy tasks  
- **Block Storage** ‚Üí best for **transactional systems** with frequent updates  
- **Object Storage** ‚Üí best for **analytics & big data** with massive scalability  

üëâ Rule of Thumb:  
- Use **File Storage** for team collaboration and shared access  
- Use **Block Storage** for **databases/VMs**  
- Use **Object Storage** for **data lakes, warehouses, ML workloads**

# Distributed Storage Systems

Distributed storage splits and replicates data across **multiple servers (nodes)**.  
Nodes form a **cluster**, giving fault tolerance, scalability, and parallel performance.

---

## üîπ How Distributed Storage Works

![Distributed Storage](./images/dds.png)

- Nodes store data on **magnetic disks** or **SSDs**.  
- Total storage = sum of all node capacities.  
- Each node can also handle replication and access control.

---

## üîπ Advantages of Distributed Storage

![Distributed Storage Advantages](./images/dds_adv.png)

- Fault tolerance and durability.  
- High availability even if some nodes fail.  
- Parallel reads/writes ‚Üí faster performance.  
- Horizontal scaling (add more nodes).

---

## üîπ Methods of Distribution

![Types of DDS](./images/types_of_dds.png)

- **Replication** ‚Üí multiple copies for availability.  
- **Partitioning (Sharding)** ‚Üí dataset split across nodes.  
- Often combined ‚Üí partition first, then replicate each shard.

---

## üîπ Replication & Partition Trade-off

![Replication & Partition](./images/dds_default.png)

- Replication = redundancy but update delay.  
- Partitioning = efficient storage spread but needs coordination.  

---

# ‚öñÔ∏è CAP Theorem in Distributed Storage

Any distributed system can only guarantee **two of three**:

- **Consistency (C)** ‚Üí latest write is always reflected.  
- **Availability (A)** ‚Üí every request gets a response.  
- **Partition Tolerance (P)** ‚Üí system works despite network failures.  

---

## üîπ Consistency

![Consistency](./images/cap_consistancy.png)  
Every read reflects the latest write.

---

## üîπ Availability

![Availability](./images/availability.png)  
Every request receives *some* response.

---

## üîπ Partition Tolerance

![Partition Tolerance](./images/partition_tolerance.png)  
System continues working even if some nodes are unreachable.

---

# üè¶ Example: Banking System (CAP Trade-off)

Two ATMs connected to the same account database:

- **CP System (Consistency + Partition Tolerance)**  
  - ATM #2 waits until ATM #1‚Äôs withdrawal is synced.  
  - ‚úÖ Balance always correct.  
  - ‚ùå Sometimes no response (reduced availability).  
  - Example: **RDBMS**.

- **AP System (Availability + Partition Tolerance)**  
  - ATM #2 shows old balance immediately.  
  - ‚úÖ Always responds fast.  
  - ‚ùå Risk of temporary inconsistency.  
  - Example: **Cassandra, DynamoDB**.

- **CA System (Consistency + Availability, no Partition Tolerance)**  
  - Only works if no network issues (rare in distributed systems).  
  - Example: **single-node DB**.

---

# üõí Example: Online Shopping (CAP Trade-off)

- **Consistency (CP)** ‚Üí System rejects your order until stock count is synced everywhere.  
- **Availability (AP)** ‚Üí Lets everyone order instantly, but may oversell stock.  

---

# üîπ ACID vs BASE in Distributed Databases

![ACID](./images/acid.png)  
![BASE](./images/base.png)  
![ACID vs BASE](./images/acid_vs_base.png)

- **ACID (Relational DBs)** ‚Üí Correctness first (Consistency + Partition Tolerance).  
- **BASE (NoSQL)** ‚Üí Scale and speed first (Availability + Partition Tolerance).
